In this notebook we're importing a datasets describing created job contracts in Lombardy (Italy). The purpose is to analyze the job market before and during the Covid emergency in order to investigate if there are categories that were more affected than others by the pandemic.

The contracts are linked to the business sector, which is the description of the corresponding "codice Ateco". For this reason we also import a dataset of such codes, which are structured in a hierarchical manner, so that we can work on more meaningful groups later.


In order to make the notebook work without changes, the dataset must be downloaded from the following links (then it may be needed to update the DATA path in the next cell, were all data files were cells are expected to be found).

```python
# Remember the trailing slash
data_path = "~/data/"
```

## Activated Jobs dataset

1. We load the datasets the with _pandas_. 
2. We convert the columns __DATA__ to the _datetime_ type.
3. We sort the DataFrame by the __DATA__ column.

#### NOTE

The _ceased_ dataset has one row with an incorrect date in the year 2600: such date is too big to be parsed correctly, so we use the option "_coerce_" to make it parse incorrect dates as _NaT_. We decided to filter out this record later since we can't be sure about what the real date is supposed to be.


```python
import pandas as pd

def prepare_dataset(path): 
    ds = pd.read_csv(path)
    ds["DATA"] = pd.to_datetime(ds["DATA"], dayfirst=True, errors="coerce")
    return ds.sort_values("DATA", ignore_index=True)

jobs = prepare_dataset(data_path + "Rapporti_di_lavoro_attivati.csv")
```

## Ateco dataset

```python
# Load dataset "Codici Ateco"
ateco = pd.read_csv(data_path + "CodiciAteco.csv")

# Separate the description from the code
ateco["Descrizione_ateco"] = ateco["Descrizione_ateco"].str.split(" - ", expand=True)[1]

# Separate the three levels of a code into separate columns
ateco[['Ateco1', 'Ateco2','Ateco3']] = ateco["Codice_ateco"].str.split(".", expand=True)

# Remove duplicates
ateco.drop_duplicates(subset=['Descrizione_ateco'], inplace=True)

# Remove useless columns
ateco.drop(["_id", "Codice_ateco"], axis=1, inplace=True)
```

## Joining into a single dataset

We concatenate the datasets into the _jobs_ dataset

```python
jobs = jobs.join(ateco.set_index("Descrizione_ateco"), on="SETTOREECONOMICODETTAGLIO")
```

## Split date into Year and Month columns

We're interested into grouping data by month so create two new columns (**ANNO** and **MESE**). Also it can be seen how all dates never go past the 12th of each months making the information about the day not relevant, in my opinion.

```python
jobs['MESE'] = jobs['DATA'].dt.month
jobs['ANNO'] = jobs['DATA'].dt.year
jobs.drop('DATA', axis=1, inplace=True)
```

## Selecting a time period
It looks like data started being recorded consistently starting by 2009 to 2021 so we'll consider this period.

#### NOTE
There is a record with an incorrect date in the _created_ dataset (2201-09-06). We decided to filter out this record  since we can't be sure about what the real date is supposed to be. Moreover the datasets have millions of records, so this would hardly make any difference.

```python
jobs.groupby(by=["ANNO", "MESE"]).size().plot()
```

```python
# Filtering by date
jobs = jobs[(jobs["ANNO"] >= 2009) & (jobs["ANNO"] <= 2021)]
jobs.groupby(by=["ANNO", "MESE"]).size().plot()
```

## Simplifying the data

By checking unique values for each column we get the following results.

1. __GENERE__: all rows are either _M_ or _F_, so nothing to do here
2. __ETA__: a few rows have really low values, like 0, 1, 2. Since these are probably errors we remove all rows with ETA < 15 from the dataset, since 15 is the legal age for working in Italy.
3. __SETTOREECONOMICODETTAGLIO__: We remove this column as we don't really need it here.
4. __TITOLOSTUDIO__: Since a few of the categories overlap it could be a good idea to group them when doing stats. Also, a big chunk of the rows have the "NESSUN TITOLO DI STUDIO", which is pretty unrealistic, so it is probably better to skip these rows when running statistics involving this column.
5. __CONTRATTO__: Also here a few values could be grouped into macro-categories. This is not a personal characteristic I want to analyze but we need this information for the numbers to make sense (renewed contracts or transformed contracts were TEMPO DETERMINATO contracts).
6. __MODALITALAVORO__: It's probably worth to group the various kind of "TEMPO PARZIALE" rows. Also we should ignore the _NON DEFINITO_ rows when doing stats on this column. There are some NaN values that will be set to _NON DEFINITO_.
7. __PROVINCIAIMPRESA__: All good here
8. __ITALIANO__: We will make this column binary (SI/NO).
9. __Ateco1, Ateco2, Ateco3__: we keep only _Codice1_ since we just want a general indication of the sector. Also we set NaN values to -1.

In the next cells we'll make these changes, by obtaining the _cleaned_jobs_ dataset.

```python
# Filtering by age
cleaned_jobs = jobs[jobs["ETA"] >= 15].copy(deep=True)

# Removing unneeded columns
cleaned_jobs.drop(['SETTOREECONOMICODETTAGLIO', 'Ateco2', 'Ateco3'], axis=1, inplace=True)

# Takes care of "grouping" categories in  a column by passing a dict of lists, 
# where the list represents the categories to replace and the key is the new name.
# When invert=False, changes the name of all values NOT in the list
#
# The order of execution of the rules is not guaranteed so it is 
# necessary to make multiple calls when that's important.
def replace_categories(dframe, column, replaceDict, invert=False):
    for k in replaceDict.keys():
        dframe.loc[dframe[column].isin(replaceDict[k])^invert, column] = k
        
# Renaming TITOLOSTUDIO to ISTRUZIONE and grouping into wider categories
cleaned_jobs.rename(columns={"TITOLOSTUDIO": "ISTRUZIONE"}, inplace=True)

replace_categories(cleaned_jobs, "ISTRUZIONE", {
    "ELEMENTARE": [
        "LICENZA ELEMENTARE"],
    "SECONDARIA INFERIORE": [
        "LICENZA MEDIA"],
    "SECONDARIA SUPERIORE": [
        "DIPLOMA DI ISTRUZIONE SECONDARIA SUPERIORE  CHE PERMETTE L'ACCESSO ALL'UNIVERSITA",
        "TITOLO DI ISTRUZIONE SECONDARIA SUPERIORE (SCOLASTICA ED EXTRA-SCOLASTICA) CHE NON PERMETTE L'ACCESSO ALL'UNIVERSITÀ ()"],
    "SCONOSCIUTO": ["NESSUN TITOLO DI STUDIO"]
})

replace_categories(cleaned_jobs, "ISTRUZIONE", {
    "TERZIARIA": [
        "ELEMENTARE",
        "SECONDARIA INFERIORE",
        "SECONDARIA SUPERIORE",
        "SCONOSCIUTO"
    ]
}, invert=True)

# Grouping CONTRATTO
replace_categories(cleaned_jobs, "CONTRATTO", {
    "TEMPO DETERMINATO": [  
        "LAVORO A TEMPO DETERMINATO", 
        "LAVORO A TEMPO DETERMINATO  PER SOSTITUZIONE",
        "CONTRATTO DI AGENZIA",
        "Lavoro a tempo determinato con piattaforma" ],
    "TEMPO INDETERMINATO": [
        "LAVORO A TEMPO INDETERMINATO", 
        "Lavoro a tempo indeterminato con piattaforma",
        "LAVORO DOMESTICO A TEMPO INDETERMINATO"],
    "TIROCINI E APPRENDISTATI": [   
        "TIROCINIO", 
        "APPRENDISTATO PROFESSIONALIZZANTE O CONTRATTO DI MESTIERE",
        "APPRENDISTATO PER LA QUALIFICA E PER IL DIPLOMA PROFESSIONALE, IL DIPLOMA DI ISTRUZIONE SECONDARIA SUPERIORE E IL CERTIFICATO DI SPECIALIZZAZIONE TECNICA SUPERIORE",
        "CONTRATTI DI BORSA LAVORO E ALTRE WORK EXPERIENCES",
        "APPRENDISTATO DI ALTA FORMAZIONE E RICERCA",
        "CONTRATTO DI FORMAZIONE LAVORO (SOLO PUBBLICA AMMINISTRAZIONE)"],
})

replace_categories(cleaned_jobs, "CONTRATTO", {
    "ALTRI": [
        "TEMPO DETERMINATO", 
        "TEMPO INDETERMINATO",
        "TIROCINI E APPRENDISTATI",]
        # "LAVORO INTERMITTENTE"]
}, invert=True)

# Grouping MODALITALAVORO and setting NaNs to "NON DEFINITO"
replace_categories(cleaned_jobs, "MODALITALAVORO", {
    "TEMPO PARZIALE": [
        "TEMPO PARZIALE ORIZZONTALE",
            "TEMPO PARZIALE MISTO",
            "TEMPO PARZIALE VERTICALE"]
})

cleaned_jobs.loc[cleaned_jobs["MODALITALAVORO"].isna(), "MODALITALAVORO"] = "NON DEFINITO"


# Making ITALIANO a boolean column
cleaned_jobs.loc[cleaned_jobs["ITALIANO"] != "ITALIA", "ITALIANO"] = "NO"
cleaned_jobs.loc[cleaned_jobs["ITALIANO"] == "ITALIA", "ITALIANO"] = "SI"

# Setting unknowns Ateco codes to -1
cleaned_jobs.loc[cleaned_jobs["Ateco1"].isna(), "Ateco1"] = -1

# Print_output
cleaned_jobs
```

## Extracting some stats

### Gender

We can see how men got more jobs during the analyzed period. From the chart we can see that this gap has widened in the latter years.

```python
import matplotlib.pyplot as plt
cleaned_jobs.groupby(["ANNO", "MESE", "GENERE"]).size().unstack().rolling(6).mean().plot()
cleaned_jobs.value_counts("GENERE")
```

### Age

It seems like there was a shift towards a higher age during the years, for both women and men. There is a more evident spike during the covid pandemic. This might be related to the fact that less new contract were made to younger people, while renewing existing contracts.

```python
cleaned_jobs.groupby(["ANNO", "MESE"]).mean().rolling(6).mean().plot()
cleaned_jobs.groupby(["ANNO", "MESE", "GENERE"]).mean().unstack().rolling(6).mean().plot()
```

## Studies

From the chart, it looks like there is a steady increasing in jobs requiring a high-level education (TERZIARIA), also it seems this category was the least affected by the Covid pandemic.

There are a lot of "SCONOSCIUTO" records. The majority of that is related to foreign workers (the second chart is only italian citizens).

```python
cleaned_jobs.groupby(["ANNO", "MESE", "ISTRUZIONE"]).size().unstack().rolling(6).mean().plot()
plt.show()
cleaned_jobs.groupby(["ANNO", "MESE", "ISTRUZIONE", "ITALIANO"]).size().unstack()["SI"].unstack().rolling(6).mean().plot()
```
